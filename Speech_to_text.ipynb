```python
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition")
```

# Building Web Application

```python
import gradio as gr
```

```python
import os
import time
```

```python
#Function to transcribe the audio

def transcribe(audio, state=" "):
  time.sleep(3)
  print(f"Received audio: {audio}")
  if not os.path.isfile(audio):
        return "Error: File does not exist", state
  try:
    text = p(audio)["text"]
  except Exception as e:
    text = f"Error: {e}"
    print(f"Exception: {e}")

  state += text + " "
  return text, state
```

```python
#
gr.Interface(
    fn=transcribe,
    inputs=[gr.Audio(type="filepath"), gr.State()],
    outputs=[gr.Textbox(), gr.State()],
    ).launch()
```
